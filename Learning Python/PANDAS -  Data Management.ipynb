# 1 what is a Dataframe?

PANDAS is a Python library that makes managing data easier and simpler than a spreadsheet.

**Wat is a Dataframe?** => It is the main structure of type of data used in PANDAS, which consists of columns and lines.

Therefore, *PANDAS is the main Python module for data analysis and management, the main structure is the Data frame, a two-dimensional rectangular arrangement, where the data is arranged in rows and columns*.

## 1.1 Convert a Dictionary into a Dataframe 

```python
#Let's check an example of a Dataframe in PANDAS

import pandas as pd #importing the PANDAS library for data management

data = {'Name' : ['Charito', 'Tina', 'Tere', 'Mila'],
         'Qualifications': ['20', '80', '50', '100'], 
         'Sports': ['Swimming ',' Baseball ',' Tennis ',' Soccer'], 
         'Subjects' : ['Mathematics', 'Statistics', 'Econometrics', 'Algebra']}

#This dataframe has three columns (Name, Ratings and Sports), and each column has 4 elements.

#We need to convert the previous Python Dictionary in a Dataframe, adding the following instruction:
df = pd.DataFrame(data)

print(df)
```

We visualize the results as if it were a spreadsheet and the columns are arranged in alphabetical order.

Important Note: *The row highlighted in orange is important, since it is the index of the DataFrame.*

# 2 Handling a Dataframe


## 2.1 Order the data 

To sort a dataframe by column, we use the function SORT.

```python
#let's run the Invoices dataframe

import pandas as pd
invoices = {'invoice': [1, 2, 3, 4, 5, 6],
            'client': [4, 1, 3, 1, 2, 6],
            'units': [3, 2, 1, 2, 1, 1],
            'price': [27.76, 21.13, 29.82, 29.96, 21.11, 23.97],
            'total': [83.28, 42.26, 29.82, 59.92, 21.11, 23.97]}

invoicesdf = pd.DataFrame(invoices)
print (invoicesdf)
```

```python
#let's check an example of ascending order

invoicesdf.sort_values(["client"], ascending=True, inplace = True)
print(invoicesdf)
```

```python
#let's check an example of descending order

invoicesdf.sort_values(["client"], ascending=False, inplace = True)
print(invoicesdf)
```

## 2.2 Add content to a dataframe 

For example, in order to add an additional column in a dataframe, we need to declare this column.

```python
#let's include an additional column
invoicesdf["Order"] = [1964375, 1964412, 1966932, 1981627, 1964416, 1961479]
print(invoicesdf)
```

```python
#In case that I want to assign the same value in the whole column:

invoicesdf["Country"] = 'ES'
print(invoicesdf)
```

## 2.3 Remove Values 

### 2.3.1 Remove values from a column

To remove the values ​​from a column, we use the **DROP function**:

```python
#Apply the DROP function
invoicesdf2 = invoicesdf.drop(['client'], axis=1)
print (invoicesdf2)
```

```python
#If you want to remove multiples columns from a dataframe, you can apply the DROP function as well.

invoicesdf3 = invoicesdf.drop(['invoice', 'client', 'units'], axis=1)
print (invoicesdf3)
```

### 2.3.2 Remove values from a line

To remove the values ​​from a line in a dataframe, we use the **DROP function**:

```python
#let's check an example with the Invoices dataframe

invoicesdf4 = invoicesdf.drop([1, 2],axis=0)
print (invoicesdf4)
```

```python
#Another way to select the rows to delete is through range, which allows you to quickly select any subset.

invoicesdf5 = invoicesdf.drop(range(1, 6, 2),axis=0)
print (invoicesdf5)
```

## 2.4 Filters and Searches 

### 2.4.1 Insert conditions in the filters 

```python
#let's apply some conditions in the filter

invoicesdf_filter = invoicesdf[ (invoicesdf['price'] > 24.00) & (invoicesdf['total'] < 30.00) ]
print(invoicesdf_filter)
```

```python
invoicesdf_byname = invoicesdf_filter[ invoicesdf_filter['Country'] == 'ES' ]
print(invoicesdf_byname)
```

### 2.4.2 Search by word length 

```python
#Build the dataframe
dfcountries = pd.DataFrame(data={"Countries":['Mexico','Argentina','Espana','Colombia'],
                        "Population":[127212000, 45167000, 47099000, 48922000]})

print(dfcountries)
```

```python
long_name = dfcountries[ dfcountries['Countries'].str.len() > 6 ]
print(long_name)
```

### 2.4.3 Search by a specific value 

```python
by_name = dfcountries[ dfcountries['Countries'] == 'Colombia' ]
print(by_name)
```

## 3.7 Series and Timeseries 

A series is a single data vector with an index that labels all elements of the vector. If we don't specify the index, an integer sequence is assigned as the index.

TimeSeries are used for the following tasks:
* Generate sequences of fixed frequency dates and time periods.
* Conform or convert time series to a particular frequency
* Calculate "relative" dates based on several non-standard time increments (for example, 5 business days before the last business day of the year), or "advance" the dates forward or backward

```python
#Let's check an example of a serie

c = pd.Series([1959, 1967, 1989, 2002])
print(c)
```

It is also possible to add tags to the indexes while creating the series.

Important note => Values ​​are stored in an array in Numpy (values), while the index is stored in a PANDAS Index object (Index).

```python
#let's check an example of a serie with an index

d = pd.Series([1959, 1967, 1989, 2002], index = ['a', 'b', 'c', 'd'])
print(d)
```

```python
#Labels can be used to refer to values ​​in the series, while we can use the index to call a specific value:

print(d['a'])
print(d[1])
```

## 3.8 Statistics 

To apply basics statistics to the dataframe, we use the **.describe() statement.**

```python
#load the csv

wine = pd.read_csv('winequality-red.csv', sep=',')
wine.head()
```

```python
#check the statistics
wine.describe()
```

To get the sum, mean and quantity of a column, we apply the following statements:

.mean() => get the mean

.sum() => get the sum

.count() => get the quantity

.min() => get the minimum value

.max() => get the maximum value


```python
#let's check an example

print(wine['quality'].mean())
print(wine['quality'].sum())
print(wine['quality'].count())
print(wine['quality'].min())
print(wine['quality'].max())
```
